```python
text_list=[]
time_list=[]
tag_list=[]
image_list=[]
for tweet in tweets:
    try:
        time=tweet.find_element(By.XPATH,'./div/div/div/div[2]/div[2]/div[1]/div/div/div[1]/a/time').get_attribute('datetime')
    except:
        time='NAN'
    try:
        tag=tweet.find_element(By.XPATH,'.//span[contains(text(),"@")]').text
    except:
        tag='NAN'
    try:
        text=tweet.find_element(By.XPATH,'.//div[@data-testid="tweetText"]/span').text
    except:
        text='NAN'
    try:
        image=tweet.find_element(By.XPATH,'.//div[@data-testid="tweetPhoto"]/img').get_attribute('src')

    if text not in text_list:
        time_list.append(time)
        tag_list.append(tag)
        text_list.append(text)
        image_list.append(image)
```
